```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
# For linear regression
import statsmodels.formula.api as sm
```

```python
#get data from the Penn World Tables excel file
pwt_df = pd.read_excel('pwt100.xlsx', 'Data')
#keep only the data we need
df = pwt_df[['countrycode', 'country', 'year', 
             'rgdpe', # Real GDP Expenditures
             'rgdpo', # Real GDP Output
             'pop',   # Population in millions
             #'hc',    # Human Capital Index
             #'ctfp',  # TFP level at current PPPs (USA=1)
             #'cwtfp'  # Welfare-relevant TFP levels at current PPPs (USA=1)
            ]]

df.to_excel("ProjectData_test.xlsx", sheet_name='ProjectData')
```

```python
#Calculate GDP per capita
for i in range(len(df.index)):
    df.loc[i,'rgdpeCapita'] = df.loc[i,'rgdpe'] / df.loc[i,'pop']
    df.loc[i,'rgdpoCapita'] = df.loc[i,'rgdpo'] / df.loc[i,'pop']

#df
```

```python
#Calculate %change YOY in  'rgdpe', 'rgdpo', 'hc', 'ctfp', 'cwtfp'
for i in range(len(df.index)):
    if df.loc[i,'year'] != 1950:
        df.loc[i,'popYOY'] = (df.loc[i,'pop'] - df.loc[(i - 1),'pop']) / df.loc[(i - 1),'pop']
        #df.loc[i,'rgdpeYOY'] = (df.loc[i,'rgdpe'] - df.loc[(i - 1),'rgdpe']) / df.loc[(i - 1),'rgdpe']
        #df.loc[i,'rgdpoYOY'] = (df.loc[i,'rgdpo'] - df.loc[(i - 1),'rgdpo']) / df.loc[(i - 1),'rgdpo']
        #df.loc[i,'hcYOY'] = (df.loc[i,'hc'] - df.loc[(i - 1),'hc']) / df.loc[(i - 1),'hc']
        #df.loc[i,'ctfpYOY'] = (df.loc[i,'ctfp'] - df.loc[(i - 1),'ctfp']) / df.loc[(i - 1),'ctfp']
        #df.loc[i,'cwtfpYOY'] = (df.loc[i,'cwtfp'] - df.loc[(i - 1),'cwtfp']) / df.loc[(i - 1),'cwtfp']
        df.loc[i,'rgdpeCapitaYOY'] = (df.loc[i,'rgdpeCapita'] - df.loc[(i - 1),'rgdpeCapita']) / df.loc[(i - 1),'rgdpeCapita']
        df.loc[i,'rgdpoCapitaYOY'] = (df.loc[i,'rgdpoCapita'] - df.loc[(i - 1),'rgdpoCapita']) / df.loc[(i - 1),'rgdpoCapita']
```

```python
# Calculate 10 year rolling averages of % YOY variables

df['popYOY10y'] = df['popYOY'].rolling(10,10,axis=0).mean(level='country')
#df['rgdpeYOY10y'] = df['rgdpeYOY'].rolling(10,10,axis=0).mean(level='country')
#df['rgdpoYOY10y'] = df['rgdpoYOY'].rolling(10,10,axis=0).mean(level='country')
#df['hcYOY10y'] = df['hcYOY'].rolling(10,10,axis=0).mean(level='country')
#df['ctfpYOY10y'] = df['ctfpYOY'].rolling(10,10,axis=0).mean(level='country')
#df['cwtfpYOY10y'] = df['cwtfpYOY'].rolling(10,10,axis=0).mean(level='country')
df['rgdpeCapitaYOY10y'] = df['rgdpeCapitaYOY'].rolling(10,10,axis=0).mean(level='country')
df['rgdpoCapitaYOY10y'] = df['rgdpoCapitaYOY'].rolling(10,10,axis=0).mean(level='country')

df.to_excel("RollingData_test.xlsx", sheet_name='RollingData')
#df
```

```python
# Calculate the number of NaNs for each country to determine which countries to use in analysis
df['rowNaN']=df.isnull().sum(axis=1) # Calculate the number of NaNs in each row
s = df.groupby('country')['rowNaN'].sum() # # Calculate the number of NaNs for each country

s = s.sort_values(ascending=True)
s.to_excel("NANCounts.xlsx", sheet_name='NANCounts')
#df
```

```python
country_list = ['Kenya','New Zealand','France','Canada','Morocco','Mexico','Mauritius','Colombia','Luxembourg',
                'Costa Rica','Japan','Cyprus','Italy','Israel','D.R. of the Congo','Denmark','Ireland','India',
                'Ecuador','Egypt','El Salvador','Iceland','Honduras','Guatemala','Ethiopia','Finland','Nicaragua',
                'Nigeria','Netherlands','Pakistan','Venezuela (Bolivarian Republic of)','Uruguay','United States',
                'United Kingdom','Argentina','Uganda','Turkey','Australia','Austria','Trinidad and Tobago','Thailand',
                'Switzerland','Sweden','Norway','Belgium','Sri Lanka','Germany','Panama','South Africa','Brazil',
                'Peru','Philippines','Bolivia (Plurinational State of)','Spain','Portugal']
country_list = sorted(country_list)
df_countries = df[df['country'].isin(country_list)]
df_countries = df_countries[df_countries['year'] >= 1960]
df_countries.to_excel("Countries.xlsx", sheet_name='Countries')
```

```python
df_1960 = df.loc[df['year']==1960]
#df_1960.rename(columns={'rgdpeCapita':'rgdpeCapita_1960'})

df_1960_rgdpeCapita = df_1960[['country', 'rgdpeCapita']]
df_1960_rgdpeCapita = df_1960_rgdpeCapita.rename(columns={'rgdpeCapita':'rgdpeCapita_1960'})

df_countries = pd.merge(df_countries, df_1960_rgdpeCapita, on = 'country')

#df_countries
df_countries.to_excel("with1960.xlsx", sheet_name='1960')
```

```python
#define data
x = df_countries.loc[df_countries['country']=='Kenya','popYOY']
y = df_countries.loc[df_countries['country']=='Kenya','rgdpeCapitaYOY']
x.dropna(inplace=True)
y.dropna(inplace=True)

#find line of best fit
a, b = np.polyfit(x, y, 1)

#add points to plot
plt.scatter(x, y)

#add line of best fit to plot
plt.plot(x, a*x+b, color = 'r')
plt.xlabel("population % change YOY")
plt.ylabel("rgdpe % change YOY")
plt.title("Kenya")
```

```python
for i in country_list:
    #define data
    x = df_countries.loc[df_countries['country']==i,'popYOY']
    y = df_countries.loc[df_countries['country']==i,'rgdpeCapitaYOY']
    x.dropna(inplace=True)
    y.dropna(inplace=True)

    #find line of best fit
    a, b = np.polyfit(x, y, 1)

    #add points to plot
    plt.scatter(x, y)

    #add line of best fit to plot
    plt.plot(x, a*x+b, color = 'r')
    plt.xlabel("popYOY")
    plt.ylabel("rgdpeYOY")
    plt.title(i)
    plt.show()
```

```python
for i in country_list:
    #define data
    x = df_countries.loc[df_countries['country']==i,'popYOY10y']
    y = df_countries.loc[df_countries['country']==i,'rgdpeCapitaYOY10y']
    x.dropna(inplace=True)
    y.dropna(inplace=True)

    #find line of best fit
    a, b = np.polyfit(x, y, 1)

    #add points to plot
    plt.scatter(x, y)

    #add line of best fit to plot
    plt.plot(x, a*x+b, color = 'r')
    plt.xlabel("popYOY 10 Year Average")
    plt.ylabel("rgdpeYOY 10 Year Average")
    plt.title(i)
    plt.show()
```

```python
regression = sm.ols(formula='rgdpeCapitaYOY10y ~ popYOY10y',data=df_countries).fit(cov_type='HC1')

print(regression.summary())
```

```python
regression = sm.ols(formula='rgdpeCapitaYOY10y ~ popYOY10y * rgdpeCapita_1960',data=df_countries).fit(cov_type='HC1')

print(regression.summary())
```

```python
regression = sm.ols(formula='rgdpeCapitaYOY10y ~ popYOY10y + rgdpeCapita_1960',data=df_countries).fit(cov_type='HC1')

print(regression.summary())
```

```python
regression = sm.ols(formula='np.log(rgdpeCapitaYOY10y+1) ~ np.log(popYOY10y+1) * np.log(rgdpeCapita_1960)',data=df_countries).fit(cov_type='HC1')

print(regression.summary())
```
